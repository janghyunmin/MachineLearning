```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import random
import math
```

```python
# Set means and Std of Payoff for the 4 Levers
mean={1:1.2,2:1.0,3:0.8,4:1.4}
sd={1:1,2:1,3:1,4:1}

np.random.seed(10)
```

```python
# Set decay rate
decay = 0.99
# Set number of iterations
nsims = 1

sum = 0
sumsq = 0

j = 1

res = pd.DataFrame()

while j<= nsims:
    qvals = {1:0, 2:0, 3:0, 4:0}
    nobs = {1:0, 2:0, 3:0, 4:0}
    totpayoff = 0

    epsilon = 1
    ntrials = 3000
    i = 1
    while i<= ntrials:
       # print(j,i)
        x = np.random.random()
        if x < epsilon:
            lever_chosen = int(4*np.random.random()) +1
            action = 'Exploration'
        else:
            lever_chosen = max(qvals, key = lambda key:qvals[key])
            action = 'Exploitation'

        lc = lever_chosen
        payoff = np.random.normal(mean[lc], sd[lc])
        nobs[lc] = nobs[lc] + 1
        qvals[lc] = qvals[lc] + (payoff-qvals[lc])/nobs[lc]
        totpayoff = totpayoff + payoff
        res0 = pd.DataFrame([{'Trial':i, 'Lever Chosen':lc, 'Action':action, 'Payoff':payoff,
                              'QVal(1)':qvals[1], 'Nobs(1)':nobs[1], 'QVal(2)':qvals[2],
                              'Nobs(2)':nobs[2], 'QVal(3)':qvals[3], 'Nobs(3)':nobs[3],
                              'QVal(4)':qvals[4], 'Nobs(4)':nobs[4], 'Average':totpayoff/i}])
        res = pd.concat([res, res0])
        i = i+1
        epsilon = epsilon*decay
    payoffpertrial = totpayoff/ntrials
    sum = sum+payoffpertrial
    sumsq = sumsq+payoffpertrial**2
    j = j +1

res
```

```python
print('decay factor= ',decay)
print('Ave Payoff = ',sum/nsims)
print('sd of payoff = ' ,math.sqrt(sumsq/nsims-(sum/nsims)**2))
```
